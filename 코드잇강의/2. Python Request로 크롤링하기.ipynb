## Request Library

```python
# request: Python으로 HTTP 요청을 보내는 라이브러리
import requests
```

### .get : 페이지 요청

```python
response= requests.get("https://www.google.com")
response
```

### .status_code : 요청 결과 (200 : 성공 / 404 : 실패 )

```python
if response.status_code == 200:
    print('Success!')
elif response.status_code == 404:
    print('Not Found.')
```

```python
type(response)
```

### .text : 내용 Text로 보기

```python
response.text[0:2000]
```

### .url : 어떤 url의 요청결과인지 보기

```python
response.url
```

### .headers : header보기

```python
response.headers
```

## 티비랭킹닷컴 시청률 크롤링

### for문 변수 출력 1. %d

```python
b=[]
for y in range(2010, 2012) :
    for m in range(1, 3) : 
        for w in range(1, 3) :
            a = "todqy is %d of %d-%d" %(w,y,m)
            b.append(a)
print(b)
```

### for문 변수출력 2. f"...{}..."  
참고: 포맷문자열리터럴 https://docs.python.org/ko/3/tutorial/inputoutput.html

```python
b=[]
for y in range(2010, 2012) :
    for m in range(1, 3) : 
        for w in range(1, 3) :
            a = f"todqy is {w} of {y}-{m}" 
            print(a)
```

```python
rating_pages = []
for y in range(2010, 2019) :
    for m in range(1, 13) : 
        for w in range(0, 5) :
            response = requests.get("https://workey.codeit.kr/ratings/index?year=%d&month=%d&weekIndex=%d" %(y,m,w))
            rating_pages.append(response.text)

print(len(rating_pages))
```
